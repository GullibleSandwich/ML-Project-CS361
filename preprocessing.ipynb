```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

import glob
import os
```

```python
csv_files = glob.glob("raw_data_1day\\" + "*.csv")

li = list()

for files in csv_files:
    df = pd.read_csv(files)
    split = files.split('_')
    i = 7
    station_name = ""
    while split[i] != 'Guwahati':
        station_name += split[i]
        i+=1
    df['Station'] = station_name
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
df.head()
```

```python
df.Station.unique()
```

```python
df.isnull().sum()
```

```python
drop_c = df.columns[df.isnull().sum() > 500]
```

```python
df.drop(drop_c, axis = 1, inplace = True)
```

```python
df.dropna(axis = 0,inplace = True)
```

```python
df
```

```python
df = df.reset_index()
df.drop('index', axis= 1,inplace=True)
```

```python
df.to_csv("clean_data\\1day_data.csv")
```
