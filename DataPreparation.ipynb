```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

import glob
import os
```

```python
def read_files(path : str):
    csv_files = glob.glob(path + "\\*.csv")

    li = list()

    for files in csv_files:
        df = pd.read_csv(files)
        split = files.split('_')
        i = 8
        station_name = ""
        while split[i] != 'Guwahati':
            station_name += split[i] + " "
            i+=1
        df['Station'] = station_name
        li.append(df)

    df = pd.concat(li, axis=0, ignore_index=True)
    return df
```

```python
df_1d = read_files("raw_data_1day")
df_15m = read_files("raw_data_15min")
```

```python
cols = df_1d.columns[0:10]
cols = cols.insert(0,df_1d.columns[-1])
cols
```

```python
def selectaqifeatures(df : pd.DataFrame):
    df = df[cols]
    df = df.reset_index()
    df.drop('index', axis= 1,inplace=True)
    return df
```

```python
## PM2.5 Sub-Index calculation
def get_PM25_subindex(x):
    if x <= 30:
        return x * 50 / 30
    elif x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0

df_15m_nan["PM2.5_SubIndex"] = df_15m_nan["PM2.5_24hr_avg"].apply(lambda x: get_PM25_subindex(x))
df_15m_clean["PM2.5_SubIndex"] = df_15m_clean["PM2.5_24hr_avg"].apply(lambda x: get_PM25_subindex(x))
```

```python
## PM10 Sub-Index calculation
def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x <= 100:
        return x
    elif x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x <= 350:
        return 200 + (x - 250)
    elif x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0

df_15m_nan["PM10_SubIndex"] = df_15m_nan["PM10_24hr_avg"].apply(lambda x: get_PM10_subindex(x))
df_15m_clean["PM10_SubIndex"] = df_15m_clean["PM10_24hr_avg"].apply(lambda x: get_PM10_subindex(x))
```

```python
## SO2 Sub-Index calculation
def get_SO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 380:
        return 100 + (x - 80) * 100 / 300
    elif x <= 800:
        return 200 + (x - 380) * 100 / 420
    elif x <= 1600:
        return 300 + (x - 800) * 100 / 800
    elif x > 1600:
        return 400 + (x - 1600) * 100 / 800
    else:
        return 0

df_15m_nan["SO2_SubIndex"] = df_15m_nan["SO2_24hr_avg"].apply(lambda x: get_SO2_subindex(x))
df_15m_clean["SO2_SubIndex"] = df_15m_clean["SO2_24hr_avg"].apply(lambda x: get_SO2_subindex(x))
```

```python
## NOx Sub-Index calculation
def get_NOx_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 180:
        return 100 + (x - 80) * 100 / 100
    elif x <= 280:
        return 200 + (x - 180) * 100 / 100
    elif x <= 400:
        return 300 + (x - 280) * 100 / 120
    elif x > 400:
        return 400 + (x - 400) * 100 / 120
    else:
        return 0

df_15m_nan["NOx_SubIndex"] = df_15m_nan["NOx_24hr_avg"].apply(lambda x: get_NOx_subindex(x))
df_15m_clean["NOx_SubIndex"] = df_15m_clean["NOx_24hr_avg"].apply(lambda x: get_NOx_subindex(x))
```

```python
## NH3 Sub-Index calculation
def get_NH3_subindex(x):
    if x <= 200:
        return x * 50 / 200
    elif x <= 400:
        return 50 + (x - 200) * 50 / 200
    elif x <= 800:
        return 100 + (x - 400) * 100 / 400
    elif x <= 1200:
        return 200 + (x - 800) * 100 / 400
    elif x <= 1800:
        return 300 + (x - 1200) * 100 / 600
    elif x > 1800:
        return 400 + (x - 1800) * 100 / 600
    else:
        return 0

df_15m_nan["NH3_SubIndex"] = df_15m_nan["NH3_24hr_avg"].apply(lambda x: get_NH3_subindex(x))
df_15m_clean["NH3_SubIndex"] = df_15m_clean["NH3_24hr_avg"].apply(lambda x: get_NH3_subindex(x))
```

```python
## CO Sub-Index calculation
def get_CO_subindex(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0

df_15m_nan["CO_SubIndex"] = df_15m_nan["CO_8hr_max"].apply(lambda x: get_CO_subindex(x))
df_15m_clean["CO_SubIndex"] = df_15m_clean["CO_8hr_max"].apply(lambda x: get_CO_subindex(x))
```

```python
## O3 Sub-Index calculation
def get_O3_subindex(x):
    if x <= 50:
        return x * 50 / 50
    elif x <= 100:
        return 50 + (x - 50) * 50 / 50
    elif x <= 168:
        return 100 + (x - 100) * 100 / 68
    elif x <= 208:
        return 200 + (x - 168) * 100 / 40
    elif x <= 748:
        return 300 + (x - 208) * 100 / 539
    elif x > 748:
        return 400 + (x - 400) * 100 / 539
    else:
        return 0

df_15m_nan["O3_SubIndex"] = df_15m_nan["O3_8hr_max"].apply(lambda x: get_O3_subindex(x))
df_15m_clean["O3_SubIndex"] = df_15m_clean["O3_8hr_max"].apply(lambda x: get_O3_subindex(x))
```

```python
# AQI bucketing function
def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN
    
df_15m_nan["Checks"] = (df_15m_nan["PM2.5_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["PM10_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["SO2_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["NOx_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["NH3_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["CO_SubIndex"] > 0).astype(int) + \
                (df_15m_nan["O3_SubIndex"] > 0).astype(int)

df_15m_nan["AQI_calculated"] = round(df_15m_nan[["PM2.5_SubIndex", "PM10_SubIndex", "SO2_SubIndex", "NOx_SubIndex",
                                 "NH3_SubIndex", "CO_SubIndex", "O3_SubIndex"]].max(axis = 1))
df_15m_nan.loc[df_15m_nan["PM2.5_SubIndex"] + df_15m_nan["PM10_SubIndex"] <= 0, "AQI_calculated"] = np.NaN
df_15m_nan.loc[df_15m_nan.Checks < 3, "AQI_calculated"] = np.NaN

df_15m_nan["AQI_bucket_calculated"] = df_15m_nan["AQI_calculated"].apply(lambda x: get_AQI_bucket(x))
df_15m_nan.head(15) # 24+3 cols
```

```python
df_15m_clean["Checks"] = (df_15m_clean["PM2.5_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["PM10_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["SO2_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["NOx_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["NH3_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["CO_SubIndex"] > 0).astype(int) + \
                (df_15m_clean["O3_SubIndex"] > 0).astype(int)

df_15m_clean["AQI_calculated"] = round(df_15m_clean[["PM2.5_SubIndex", "PM10_SubIndex", "SO2_SubIndex", "NOx_SubIndex",
                                 "NH3_SubIndex", "CO_SubIndex", "O3_SubIndex"]].max(axis = 1))
df_15m_clean.loc[df_15m_clean["PM2.5_SubIndex"] + df_15m_clean["PM10_SubIndex"] <= 0, "AQI_calculated"] = np.NaN
df_15m_clean.loc[df_15m_clean.Checks < 3, "AQI_calculated"] = np.NaN

df_15m_clean["AQI_bucket_calculated"] = df_15m_clean["AQI_calculated"].apply(lambda x: get_AQI_bucket(x))
df_15m_clean.head(15) # 24+3 cols
```
